Neste exemplo mostra-se como obter as cotações de uma ação para um mês e mostrar cálculos e gráficos sobre a mesma, utilizando para isso as diferentes bibliotecas do Python.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn
import urllib.request
%matplotlib inline
```

```python
#Caso queira ir buscar os dados à internet terá de passar um URL válido
#urllib.request.urlretrieve(
#    'https://query1.finance.yahoo.com/v7/finance/download/BCP.LS?period1=1493348019&period2=1524884019&interval=1d&events=history&crumb=AjVhRK.0yPw',
#    'bcp2.csv'
#)

open('BCP.LS.csv','r').readlines()[:10]

df_bcp = pd.read_csv('BCP.LS.csv',parse_dates=['Date'])

print(df_bcp.head())

print(df_bcp['Close'].describe())

df_bcp = df_bcp.sort_values(by='Date')

df_bcp.set_index('Date',inplace=True)

df_bcp['Close'].plot(figsize=(10, 8))

plt.show()

df_bcp.truncate(before='2018-01-01')['Close'].plot(figsize=(10, 8))

plt.show()

df_bcp['Close'].plot(figsize=(10, 8))

df_bcp.truncate(before='2018-01-01')['Close'].plot(figsize=(10, 8))

plt.show()

# Calcular as medias moveis de 20, 50 e 100 dias
bcp_ma20 = df_bcp['Close'].rolling(window=20).mean()
bcp_ma50 = df_bcp['Close'].rolling(window=50).mean()
bcp_ma100 = df_bcp['Close'].rolling(window=100).mean()

# Desenhar o grafico com as medias moveis
fig, ax = plt.subplots(figsize=(10,8))

ax.plot(df_bcp.index, df_bcp['Close'], label='BCP')
ax.plot(bcp_ma20.index, bcp_ma20, label='ma 20 dias')
ax.plot(bcp_ma50.index, bcp_ma50, label='ma 50 dias')
ax.plot(bcp_ma100.index, bcp_ma100, label='ma 100 dias')

ax.set_xlabel('Data')
ax.set_ylabel('Preco do fecho')
ax.legend()

plt.show()
```
