**Função Lambda**

Pequenas funções anónimas que podem ter vários argumentos mas somente uma expressão.

`````python
#Somar tres numeros
soma_tres = lambda a,b,c: print(a+b+c)
soma_tres(4,5,1)

import math
#Função que calcula o valor da hipotenusa
#h = raiz (c1**2 + c2**3)
hipotenusa = lambda c1,c2: math.sqrt(c1*c1+c2*c2)
hipotenusa(3,4)
`````

`````python
#Descobrir o maior elemento
max_tres = lambda a,b,c: a if a>b and b>c else (b if b>a and b>c else c)
print(max_tres(6,15,1))
`````

**Função Map**

Atribui a cada elemento de uma lista uma função

`````python
#Calcular os quadrados dos elementos de uma lista
#Exemplo 1
nums = [1,2,3,4,5]
def sq(n):
  return n*n
square = list(map(sq,nums))
print(square)
`````

`````python
#Exemplo 2 - utilizando uma função lambda
nums = [1,2,3,4,5]
square = list(map(lambda x: x*x,nums))
print(square)
`````

`````python
#Exemplo 3 - retirar os primeiros nomes de uma lista de nomes
names = [
    {'first': 'lokesh', 'last': 'sharma'},
    {'first': 'Astha', 'last': 'verma'},
    {'first': 'jiu', 'last': 'rai'}
 ]
first_names = list(map(lambda x: x['first'], names))
print(first_names)
`````

`````python
#Comparação entre map e list comprehension
xs=range(10)
%timeit map(lambda x: x+2, xs)
%timeit [x+2 for x in xs]
`````

**Função Filter**

Função que filtra os elementos testando se a condição é verdadeira ou falsa

`````python
#Exemplo 4 - filtrar os números pares/impares ed uma lista
seq = [0, 1, 2, 3, 4, 5]

# obtem os numeros impares da lista
result = filter(lambda x: x % 2 == 1, seq)
print(list(result))

# obtem os numeros pares da lista
result = filter(lambda x: x % 2 == 0, seq)
print(list(result))
`````

**Função Reduce**

Esta função aceita uma função e uma lista e devolve um único valor calculado.


```
A função reduce:

functools.reduce(function, iterable[, initializer])

Equivale a:

def reduce(function, iterable, initializer=None):
    it = iter(iterable)
    if initializer is None:
        value = next(it)
    else:
        value = initializer
    for element in it:
        value = function(value, element)
    return value

````

`````python
from functools import reduce
seq=[2,3,4,5,6]
somar_seq=reduce(lambda a,b:a+b,seq)
print(somar_seq)
`````

**Utilização das funções Map Reduce**

`````python
from functools import reduce

numeros = [1,2,3,4,5]
results1 = list(map(lambda x: x+2, numeros))
print(results1)

results2 = reduce(lambda x,y: x+y, results1)
print(results2)

print(reduce(lambda x,y: x+y, list(map(lambda x: x+2, numeros))))
`````

`````python
# Python 3
my_strings = ['a', 'b', 'a', 'b', 'e']
my_numbers = [1,2,3,4,5]

results1 = list(map(lambda x, y: (x, y), my_strings, my_numbers))
print(results1)

results2 = reduce(lambda x , y: (x[0] + y[0], x[1] + y[1]), results1)
print(results2)
`````

**Função Zip**

Esta função recebe coleções enumeráveis (iterables) e cria um iterador com os elementos das coleções agregados.

1.   Se não forem passads coleções devolve um iterador vazio.
2.   Se for passado uma coleção devolve uma coleção com 1 tuplo.
3.   Se for pasado mais que uma coleção é  devolvido 1 tuplo com o tamanho da coleção mais pequeno.



`````python
#Exemplo 5 - juntar os elementos das três listas
name = ["Manjeet", "Nikhil", "Shambhavi"]
roll_no = [4, 1, 3]
marks = [40, 50, 60]

mapped = zip(name, roll_no, marks)

print(list(mapped))
`````
