```python
#Esta parte rebenta no google drive - tem de ser com menos ficheiros
from datetime import datetime
from glob import glob
import pandas as pd

#all_files = glob('yellow_tripdata_2019-*.csv')
#li = []
#for filename in all_files:
#    df_temp = pd.read_csv(filename, index_col=None, header=0)
#    li.append(df_temp)
#df = pd.concat(li, axis=0, ignore_index=True)

df = pd.concat(map(pd.read_csv, glob('yellow_tripdata_2019-*.csv')))
start = datetime.now()
res = df.groupby("DOLocationID")["fare_amount"].sum().sort_values(ascending=False).head(10)
print(res)
print(f"Runtime: {datetime.now() - start}")
```

```python
from datetime import datetime
import pyspark.pandas as ps 

pdf = ps.read_csv("yellow_tripdata_2019-*.csv")
print(pdf.count())
start = datetime.now()
res = pdf.groupby("DOLocationID")["fare_amount"].sum().sort_values(ascending=False).head(10)
print(res)
print(f"Runtime: {datetime.now() - start}")
```
