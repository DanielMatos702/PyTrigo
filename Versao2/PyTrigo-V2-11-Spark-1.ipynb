# Exemplo helloword

Correr exemplo helloworld do spark utilizando **Spark Context** que consite na criação de uma aplicação que vai ler para rdds diferentes cada linha de texto de um ficheiro de texto (há outros métodos para ler o ficheiro todo ou ler vários ficheiros de uma vez, isto é só um exemplo).

```python
#1. Criar o ficheiro
!echo -e "ola bom dia \nola bom dia2 \nlkl lkl \nooi oop ppoi \npython ppp \nijij python" > testfile1
```

```python
!more testfile1
```

```python
#2. Criar a sessão spark
#Com instalação manual
#import findspark
#findspark.init()

import pyspark
#local[*]->significa que estamos a correr num cluster local
sc = pyspark.SparkContext('local[*]')
#Caso a sessão já esteja criada pois podem já ter sorrido esta linha utilizem a
#seguinte instrução:
#sc = SparkContext.getOrCreate('local[*]')
```

```python
#3. Ler o ficheiro
#Contar as linhas
txt = sc.textFile('testfile1')
print(txt.collect())
print(txt.count())
#Contar as linhas que contêm a palavra python
python_lines = txt.filter(lambda line: 'python' in line.lower())
print(python_lines.count())
```

```python
#4. Fechar o contexto - fazer isto se não se pretender continuar a utilizar o mesmo contexto
#sc.stop()
```

```python
#Exemplo 2
#Pegar numa lista de inteiros, carregar num RDD e utilizar 
#uma função lambda para retirar os números pares
#sc = SparkContext.getOrCreate('local[*]')
lista = range(10000)
rdd = sc.parallelize(lista, 2)
impares = rdd.filter(lambda x: x % 2 != 0)
print(impares.take(10))
#sc.stop()
```

```python
#Exemplo 3
#Criar uma lista aleatoria de numeros e mostrar algumas estatisticas
import numpy as np

lista = np.random.randint(0, 10, 20)
rdd = sc.parallelize(lista, 2)
print(rdd.count())
print(rdd.first())
print(rdd.min())
print(rdd.max())
print(rdd.collect())
```

```python
rdd = spark.sparkContext.textFile('pg100.txt')
```

```python
#Para mostrar os dados todos deve ser utilizado a função collect()
#rdd.collect()
rdd.take(10)
```

```python
#2 Retira linhas sem conteúdo
linhas_com_texto = rdd.filter(lambda x: len(x) > 0)
linhas_com_texto.take(5)
```

```python
#3 Separa todas as palavras
palavras = linhas_com_texto.flatMap(lambda x: x.split(' '))
palavras.take(5)
```

```python
#4 Conta as palavras existentes
#4.1 As palavras são todas mapeadas para ter o valor 1
#4.2 As palavras são agrupadas por palavra e somadas as suas ocorrências (reduceByKey)
contar_palavras = palavras.map(lambda x:(x,1)).reduceByKey(lambda x,y: x+y).map(lambda x: (x[1],x[0])).sortByKey(False)
```

```python
#5 Mostra as palavras
#Se quiserem correr para todo o rdd, substituam take(10) -> collect()
print(contar_palavras.take(10))
```

```python
#6 Guardar o resultado para ficheiro
#sc.parallelize(range(10)).saveAsTextFile('asdsd')
#Verificar: !more asdsd/part-00000*
contar_palavras.saveAsTextFile('output')
#Verificar
#from fileinput import input
#from glob import glob
#''.join(sorted(input(glob('resultado1.txt' + "/part-0000*"))))
```
