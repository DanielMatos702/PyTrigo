**Tipos de dados**

Em Python tudo são objetos e todos os objetos são de um determinado tipo.

O Python trabalha com os seguintes tipos de dados básicos: 
  - int (números inteiros)
  - float (números reais)
  - str (texto)
  - bool (booleanos - verdadeiro (True) ou falso (False))
  - NoneType (tipo especial que representa a situação sem valor)

Em Python não é preciso declarar de forma explicita o tipo de dados, ele próprio deduz a partir da utilização que lhe estamos a dar (Duck Typing). Em casos de desambiguação é necessário indicar o tipo de dados.

**Repetições ou ciclos (for e while)**

As instruções de repetição permitem repetir uma ou mais instruções até uma determinada condição se verificar.

As instruções que se utilizam para fazer repetições são:

```
for variavel in qualquercoisa:
  instrucoes
while condicao:  
  instrucoes
```

No caso do *for*, dependendo da situação em causa, é necessário indicar o intervalo de valores que se quer percorrer. Para tal utiliza-se a função/método range() com diferentes chamadas:

* range(n) -> 0, 1, 2, 3, ..., n-1 (percorre todos os valores de 0 a n-1 andando de 1 em 1 (este andar de 1 em 1 é designado de incemento (step)))
* range(m,n) -> m, m + 1, m + 2, ..., m + n - 1 (igual ao anterior, mas entre dois valores definidos pelo utilizador)
* range(m,n,s) -> m, m + s, m + 2s, ..., m + ((m + n - 1)/s) * s (aqui para além do limite inferior e superior também se pode definir o incremento (step), ou seja, de quanto em quanto é que são gerados os números, pro exemplo, para um step = 2, os números são gerados de 2 em 2)

As instruções continue e break permitem interromper a normal execução do ciclo. A primeira salta para a próxima iteração, já a segunda interrompe a execução do ciclo.

## Métodos (funções e procedimentos)

Uma função é um bloco de código que executa um conjunto de instruções. 

As funções comportam-se como pequenos programas, podendo ter entradas e saídas. As entradas são os parâmetros das mesmas, por exemplo, para calcular o seno de um angulo temos que passar à função que faz esse cálculo o angulo em questão (ex.: sin(alpha)). As saídas são os valores devolvidos pela função. No caso anterior o valor do seno do angulo. No Python as funções podem devolver mais que um valor, separados por vírgulas, sob a forma de um tuplo.

As funções são definidas com a palavra **def**.

A sintaxe é:

```python
def nomedometodo(arg1, arg2,... argN):
    ''' Document String'''
    instrucoes
    return <valores>```
    
Permitem decompor o programa em partes mais pequenas.

Dado que o Python é uma linguagem orientada a objetos, as funções são designadas de métodos. 

Ao longo destes textos dependendo da situação utilizar-se-á a palavra função ou método.

````python
#Mostra o fatorial de um numero
#5!=5x4x3x2x1
#def fatorial(numero):
#  for i in range(1,numero+1)
def fatorial(numero):
  res = 1
  for i in range(numero,0,-1):
    res = res * i
  return res

print("O fatorial de 5 e %d" % fatorial(5))
````

````python
#Mostra o fatorial de um numero
#5!=5x4x3x2x1
def fatorial2(numero):
  res = 1
  i = numero
  while (i>0):
    res = res * i
    i=i-1
  return res

print("O fatorial de 5 e %d" % fatorial2(5))
````

````python
#Mostre um fatorial de um numero de forma recursiva
#5!=5x4x3x2x1
#4!=4x3x2x1
#5!=5x4!
#Fatorial recursivo
def factrec(numero):
  #5!=5x4!
  if numero == 1:
    return 1
  else:
    return numero*factrec(numero-1)

print(factrec(5))
````

````python
#Ex29: Sequência Fibonacci (forma recursiva)
# 1 + 1 + 2 + 3 + 5 + 8 (Soma dos dois anteriores)
# 0   1   2   3   4   5  
# Fib(5) = 8 -> Fib(4) + Fib(3)
# Fib(4) = 5
# Fib(3) = 3

def fibo(a):
  if a>1:
    return fibo(a-1) + fibo(a-2)
  else:
    return 1

print(fibo(6))

#6? -> 13
````
